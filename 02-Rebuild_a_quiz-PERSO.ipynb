```python
def quizz():
    nb_de_chances = 3
    questions = {
        "Combien de fois la France a gagné la coupe du monde ? ": 2,
        "Quand a été fondé Apple ? ": 1976,
        "Qui a fondé SpaceX ? ": "elon musk"
    }    

    for question, right_answer in questions.items():
        
        while nb_de_chances > 0:
            user_answer = input(question)
            
            if isinstance(right_answer, int): # Conversion en int si user input nombre
                user_answer = int(user_answer)
            else:
                user_answer = user_answer.lower() # Juste un lower_casing si c'est un str (val par def des inputs)
            
            if user_answer == right_answer:
                print("Bonne réponse !")
                break
            else:
                nb_de_chances -= 1
                if nb_de_chances > 0:
                    print(f"Dommage ! Il te reste {nb_de_chances} chances")
        
        # Si on a plus de chances, on arrête tout
        if nb_de_chances == 0:
            break
    
    # Un seul endroit pour les messages finaux
    if nb_de_chances == 0:
        print("Oh non ! Tu as perdu le jeu...")
    else:
        print("Bravo ! Tu as réussi le quiz !")
```
